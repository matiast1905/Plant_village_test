# Import required libraries and data

```python
import tensorflow as tf
from tensorflow.keras import models, layers
import matplotlib.pyplot as plt
import os
os.chdir('/content/drive/MyDrive/Kaggle/plant_village/Potato_test')
```

```python
dataset = tf.keras.preprocessing.image_dataset_from_directory(
    'data',
    seed=42,
    image_size = (256,256),
    batch_size = 3,
    shuffle = True
)
```

```python
names = dataset.class_names
```

```python
names
```

# Plot full dataset

```python
plt.figure(figsize=(20,15))
for index,(image_batch, labels_batch) in enumerate(dataset):
    for i in range(3):
        plt.subplot(4, 3, i + 1+3*index)
        plt.imshow(image_batch[i].numpy().astype("uint8"))
        plt.title(names[labels_batch[i]]+f' batch {index}')
        plt.axis("off")
```

# Divide data into train and test

```python
train = dataset.take(3)
test = dataset.skip(3)
```

# Plot train dataset

```python
plt.figure(figsize=(20,15))
for index,(image_batch, labels_batch) in enumerate(train):
    for i in range(3):
        plt.subplot(4, 3, i + 1+3*index)
        plt.imshow(image_batch[i].numpy().astype("uint8"))
        plt.title(names[labels_batch[i]]+f' train batch {index}')
        plt.axis("off")
```

# Plot test dataset

```python
plt.figure(figsize=(20,15))
for index,(image_batch, labels_batch) in enumerate(test):
    for i in range(3):
        plt.subplot(4, 3, i + 1+3*index)
        plt.imshow(image_batch[i].numpy().astype("uint8"))
        plt.title(names[labels_batch[i]]+f' test batch {index}')
        plt.axis("off")
```
